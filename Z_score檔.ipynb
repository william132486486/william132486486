```python
A_1103 = (df1.loc['1103 CHC','Total current assets'] - df1.loc['1103 CHC','Total current liabilities'])/df1.loc['1103 CHC','Total assets']
```

```python
X1 ='Total current assets'
X2 ='Total current liabilities'
X3 ='Total assets'
X4 ='Total retained earnings'
X5 ='Earnings before interest and tax'
X6 ='Total equity'
X7 ='Total liabilities'
X8 ='Total operating revenue'
CHC ='1103 CHC'
UCC ='1104 UCC'
```

```python
A_CHC = (df1.loc[CHC,X1] - df1.loc[CHC,X2])/df1.loc[CHC,X3]
```

```python
B_CHC = df1.loc[CHC,X4]/df1.loc[CHC, X3]
```

```python
C_CHC = df2.loc[CHC,X5]/df1.loc[CHC, X3]
```

```python
D_CHC = df1.loc[CHC,X6]/df1.loc[CHC, X7]  #調整分子項定義，採用股東權益帳面價值
```

```python
E_CHC = df2.loc[CHC,X8]/df1.loc[CHC, X3]
```

```python
Z_score_CHC = 1.2*A_CHC + 1.4*B_CHC + 3.3*C_CHC + 0.6*D_CHC+1.0*E_CHC
```

```python
Z_score_CHC, A_CHC, B_CHC, C_CHC, D_CHC, E_CHC
```
